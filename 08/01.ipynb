```python
import cv2 
from skimage.feature import graycomatrix, graycoprops
import numpy as np
import matplotlib as plt
```

```python
# Baca citra biner (hasil segmentasi)
img = cv2.imread("cat.png", 0)
_, thresh = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)

# Temukan kontur
contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

for cnt in contours:
    area = cv2.contourArea(cnt)
    perimeter = cv2.arcLength(cnt, True)
    x, y, w, h = cv2.boundingRect(cnt)
    aspect_ratio = float(w) / h

    print(f"Luas: {area},Keliling: {perimeter}, Aspect Ratio: {aspect_ratio:.2f}")
```

```python
# Hitung GLCM
glcm = graycomatrix(
    img, distances=[1], angles=[0], levels=256, symmetric=True, normed=True
)

# Ekstraksi properti tekstur
contrast = graycoprops(glcm, "contrast")[0, 0]
homogeneity = graycoprops(glcm, "homogeneity")[0, 0]
energy = graycoprops(glcm, "energy")[0, 0]
correlation = graycoprops(glcm, "correlation")[0, 0]

print(f"Contrast: {contrast:.2f}")
print(f"Homogeneity: {homogeneity:.2f}")
print(f"Energy: {energy:.2f}")
print(f"Correlation: {correlation:.2f}")
```

```python
# Baca Citra Warna
img = cv2.imread('cat.png')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

#Rata-rata nilai warna per channel
mean_r = np.mean(img_rgb[:,:,0])
mean_g = np.mean(img_rgb[:,:,1])
mean_b = np.mean(img_rgb[:,:,2])

print(f'Rata-rata R: {mean_r:.2f}, G: {mean_g:.2f}, B: {mean_b:.2f}')

#Histogram Warna
colors = ('r','g','b')
for i, col in enumerate(colors):
    hist = cv2.calcHist([img_rgb],[i], None, [256],[0, 256])
    plt.plot(hist, color=col)
    
plt.title('Histogram Warna')
plt.show()
```
