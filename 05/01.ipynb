```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('img.jpg')

#Mean Filter
mean = cv2.blur(img,(5,5))

#Gaussian Filter
gauss = cv2.GaussianBlur(img, (5,5), 0)

#Median Filter
median = cv2.medianBlur(img,5)
```

```python
plt.figure(figsize=(10,4))

plt.subplot(1,4,1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original')
plt.axis('off')

plt.subplot(1,4,2)
plt.imshow(cv2.cvtColor(mean, cv2.COLOR_BGR2RGB))
plt.title('Mean')
plt.axis('off')

plt.subplot(1,4,3)
plt.imshow(cv2.cvtColor(gauss, cv2.COLOR_BGR2RGB))
plt.title('Gaussian')
plt.axis('off')

plt.subplot(1,4,4)
plt.imshow(cv2.cvtColor(median, cv2.COLOR_BGR2RGB))
plt.title('Median')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
img2 = cv2.imread('../cat.png')

# Kernel penajaman sederhana
kernel = np.array([[0,-1,0],[-1,5,-1],[0,-1,0]])

# Terapkan kernel ke citra
sharpened = cv2.filter2D(img2,-1,kernel)
sharpened_rgb = cv2.cvtColor(sharpened, cv2.COLOR_BGR2RGB)

```

```python
plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
plt.imshow(img2)
plt.title('Asli')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(sharpened_rgb)
plt.title('Hasil Sharpening')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
# Membaca Citra Biner
img = cv2. imread('biner.png', 0)
_, binary = cv2.threshold(img, 127 , 255 , cv2.THRESH_BINARY)
```

```python
# Structuring element
kernel = np.ones((5,5),np.uint8)

#Operasi erosi dan dilasi
eros = cv2.erode(binary, kernel, iterations=2)
dil= cv2.dilate(binary, kernel, iterations=2)
```

```python
plt.subplot(1,3,1)
plt.imshow(binary, cmap='gray')
plt.title('Original')
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(eros, cmap='gray')
plt.title('Erosi')
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(dil, cmap='gray')
plt.title('Dilasi')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
# Opening dan Closing
opening = cv2.morphologyEx(binary, cv2.MORPH_OPEN, kernel)
closing = cv2.morphologyEx(binary, cv2.MORPH_CLOSE, kernel)
```

```python
# Tampilkan Hasil
plt.subplot(1,3,1)
plt.imshow(binary, cmap='gray')
plt.title('Original')
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(opening, cmap='gray')
plt.title('Opening')
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(closing, cmap='gray')
plt.title('Closing')
plt.axis('off')

plt.tight_layout()
plt.show()

```
