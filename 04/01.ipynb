```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from skimage import exposure
```

```python
img = cv2.imread('gambar.jpg',cv2.IMREAD_GRAYSCALE)
```

```python
plt.subplot(1,2,1)
plt.imshow(img, cmap='gray')
plt.title('Gambar Asli')
plt.axis('off')

plt.subplot(1,2,2)
plt.hist(img.ravel(),256,[0, 256])
plt.title('Histogram')
plt.xlabel('Intensitas Piksel')
plt.ylabel('Jumlah')

plt.tight_layout()
plt.show()
```

```python
equalized = cv2.equalizeHist(img)
```

```python
plt.subplot(1,2,1)
plt.imshow(img, cmap='gray')
plt.title('Gambar Asli')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(equalized, cmap='gray')
plt.title('Setelah Equalized')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
source = cv2.imread('gambar.jpg',cv2.IMREAD_GRAYSCALE)
reference = cv2.imread('gambar.jpg',cv2.IMREAD_GRAYSCALE)
matched = exposure.match_histograms(source, reference, channel_axis=None)
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,3,1)
plt.imshow(source, cmap='gray')
plt.title('Sumber')
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(reference, cmap='gray')
plt.title('Referensi')
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(matched, cmap='gray')
plt.title('Hasil Matching')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
p_min, p_max = np.percentile(img, (0, 100))
stretched = np.clip((img - p_min) * 255.0 / (p_max - p_min), 0, 255).astype(np.uint8)
```

```python
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Gambar Asli')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(stretched, cmap='gray')
plt.title('Setelah Contrast Stretching')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
gamma = 2
gamma_corrected = np.power(img / 255.0, gamma) * 255
gamma_corrected = np.clip(gamma_corrected, 0, 255).astype(np.uint8)
```

```python
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Gambar Asli')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(gamma_corrected, cmap='gray')
plt.title('Setelah Gamma Correction')
plt.axis('off')

plt.tight_layout()
plt.show()
```
